## 1. Apresentação de casos nominais e de falha em períodos próximos:

### a)      21/04/2017 -> Poço 1 

```python
data_nom = pd.read_csv('3W/0/WELL-00001_20170421100251.csv')
data_nom = data_nom.drop(columns=['T-JUS-CKGL','QGL'])
data_nom.head()
```

```python
data_1 = pd.read_csv('3W/4/WELL-00001_20170421160000.csv')
data = data_1.drop(columns=['T-JUS-CKGL','QGL'])
data.head()
```

```python
for i in data.columns[1:]:
    plt.figure(figsize=(6,4))
    plt.plot(data[i], label = 'Fault')
    plt.plot(data_nom[i], label = 'Nominal')
    plt.legend()
    plt.title(i)
    plt.xlabel('Time')
    plt.ylabel('Pressure')
    
```

### c) 06/08/2014 -> Poço 4 -> Transição entre falha e caso nominal:

```python
# caso nominal
data_nom_w3 = pd.read_csv('3W/0/WELL-00004_20140806220033.csv')
data_nom_w3 = data_nom_w3.drop(columns=['P-JUS-CKGL','T-JUS-CKGL','QGL'])
# caso com falha
data_fault_w3 = pd.read_csv('3W/4/WELL-00002_20131214220010.csv') 
data_fault_w3 = data_fault_w3.drop(columns=['P-JUS-CKGL','T-JUS-CKGL','QGL'])

data_f_w3 = np.vstack((data_fault_w3.values,data_nom_w3.values))
dados_w3 = pd.DataFrame(data_f_w3, columns=data_nom.columns)
```

```python

for i in data_nom_w3.columns[1:]:
    plt.figure(figsize=(12,4))
    plt.plot(dados_w3[i], 'k', label = 'Nominal')
    plt.plot(data_fault_w3[i], 'r', label = 'Falha')
    plt.legend()
    plt.title(i)
    plt.xlabel('Time')
    plt.ylabel('Pressure')
```
